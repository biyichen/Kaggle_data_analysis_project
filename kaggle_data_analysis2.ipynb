```python
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# turn all categorical variables into numerical variables
cat_train = train.select_dtypes(['object']).columns
train[cat_train] = train[cat_train].apply(lambda x: x.astype('category'))
train[cat_train] = train[cat_train].apply(lambda x: x.cat.codes)
train
```

```python
# split X_train set and y_train set
y_train = train['loss']
X_train = train.drop(['loss'], axis=1)
```

```python
# check categorical variables and numerical valuables
test.dtypes
```

```python
# select categorical variables
cat_test = test.select_dtypes(include=['object']).copy()
cat_test.head()
```

```python
# check missing values
test[test.isnull().any(axis=1)]
# found that there is no missing values in categorcial variables
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# turn all categorical variables into numerical variables
cat_test = test.select_dtypes(['object']).columns
test[cat_test] = test[cat_test].apply(lambda x: x.astype('category'))
test[cat_test] = test[cat_test].apply(lambda x: x.cat.codes)
test
```

```python
from sklearn import linear_model
import numpy as np

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Evaluation metrics on test data
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
 % np.mean((regr.predict(X_train) - y_train) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X_train, y_train))
```

```python
predictions = regr.predict(test)
predictions
```

```python
# CRATE YOUR SUBMISSION FILE HERE
import pandas as pd

# create a submission for Kaggle
submission = pd.DataFrame({"id":test["id"],
                          "loss":predictions})
# save submission to CSV
submission.to_csv("sample_submission.csv", index=False)
```
